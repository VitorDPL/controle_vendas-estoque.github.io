```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('vendas.csv')
```

```python
df= df.rename(columns= {"number" : "numeração"})
df= df.rename(columns= {"quantidade_pecas": "quantidade"})
df= df.rename(columns= {'tipo_peca' : 'Tipo de roupa'})
display(df)
```

### Numeração mais vendida

```python
df_number= df[[ 'numeração', 'quantidade' ]]
df_number= df_number.groupby('numeração').sum('quantidade')

#display(df_number)

df_number.sort_values(by='quantidade', ascending=False)

display(df_number)
```

### Referência mais vendida

```python
df_ref= df.groupby('ref').sum('quantidade')
#display(df_ref)
df_ref= df_ref.drop(['numeração'], axis=1)
df_ref= df_ref.drop(['id'], axis=1)

df_ref.sort_values(by='quantidade', ascending=False)
```

### Tipo de roupa mais vendido:


```python
df_tipo= df.groupby('Tipo de roupa').sum('quantidade')
df_tipo= df_tipo[['quantidade']].sort_values(by='quantidade', ascending= False)
display(df_tipo)
```
