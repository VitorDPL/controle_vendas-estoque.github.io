```python
import pandas as pd
```

```python
df= pd.read_csv('estoque.csv')
display(df)
```

```python
#Quantidade total de peças;

df_soma_das_pecas= df['quantidade_pecas'].sum()
print(df_soma_das_pecas)
```

### Quantidade de peças por referência:

```python
df_ref= df[['ref', 'tipo_peca' , 'quantidade_pecas']]

df_ref= df_ref.groupby("ref").sum('quantidade_pecas')

display(df_ref)
```

### Quantidade de peças por tamanho:

```python
df_tam= df[['tamanho', 'quantidade_pecas']]

df_tam= df_tam.groupby('tamanho').sum('quantidade_pecas')

conta_percentual = ((df_tam['quantidade_pecas'] * 100) / df_soma_das_pecas).astype(int)

#conta_percentual = conta_percentual.astype(int)



#print(conta_percentual)

df_tam['percentual'] = conta_percentual

display(df_tam)
```

### Quantidade de peças por tipo.


```python
df_tipo= df[['tipo_peca', 'quantidade_pecas']]
df_tipo= df_tipo.groupby("tipo_peca").sum("quantidade_pecas")
display(df_tipo)
```
